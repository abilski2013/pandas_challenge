```python
import pandas as pd
```

```python
file = "Resources/schools_complete.csv"
schools_complete = pd.read_csv(file)
file2 = "Resources/students_complete.csv"
students_complete = pd.read_csv(file2)
schools_complete.head()
#students_complete.head()


```

```python
### District Summary

#* Create a high level snapshot (in table form) of the district's key metrics, including:
 # * Total Schools
 # * Total Students
 # * Total Budget
 # * Average Math Score
 # * Average Reading Score
 # * % Passing Math  NEED 70% minimum to pass
 # * % Passing Reading  NEED 70% minimum to pass
 # * Overall Passing Rate (Average of the above two)
total_schools = schools_complete["school_name"].count()
total_students = schools_complete["size"].sum()
total_budget = schools_complete["budget"].sum()
avg_math = students_complete["math_score"].mean()
avg_reading = students_complete["reading_score"].mean()
total_students = students_complete["Student ID"].count()
math_passing_only = students_complete.loc[students_complete["math_score"]>=70]
passing_math_count = math_passing_only["Student ID"].count()
percent_passing_math = passing_math_count/total_students
reading_passing_only = students_complete.loc[students_complete["reading_score"]>=70]
passing_reading_count = reading_passing_only["Student ID"].count()
percent_passing_reading = passing_reading_count/total_students
overall_passing_rate = (percent_passing_math + percent_passing_reading)/2

Summary_Data_Table = pd.DataFrame({"Total Schools" : [total_schools], "Total Students" : [total_students],
                                   "Total Budget" : [total_budget], "Avg Math Score" : [avg_math],
                                   "Avg Reading Score" : [avg_reading], "% Passing Math" : [percent_passing_math],
                                   "% Passing Reading" : [percent_passing_reading],
                                   "Overall Passing Rate" : [overall_passing_rate]})

Summary_Data_Table["Total Budget"] = Summary_Data_Table["Total Budget"].map("${:.2f}".format)
Summary_Data_Table["% Passing Math"] = Summary_Data_Table["% Passing Math"].map("{:.2%}".format)
Summary_Data_Table["% Passing Reading"] = Summary_Data_Table["% Passing Reading"].map("{:.2%}".format)
Summary_Data_Table["Overall Passing Rate"] = Summary_Data_Table["Overall Passing Rate"].map("{:.2%}".format)

Summary_Data_Table
```

```python
### School Summary

#* Create an overview table that summarizes key metrics about each school, including:
#  * School Name
#  * School Type
#  * Total Students
#  * Total School Budget
#  * Per Student Budget
#  * Average Math Score
#  * Average Reading Score
#  * % Passing Math
#  * % Passing Reading
#  * Overall Passing Rate (Average of the above two)

schools_complete["budget_per_student"] = schools_complete["budget"]/schools_complete["size"]
schools_complete
avg_math_by_school = students_complete.groupby(["school_name"])
avg_reading_by_school = students_complete.groupby(["school_name"])
merge_table = pd.merge(schools_complete, avg_math_by_school["math_score"].mean(), on="school_name")
merge_table = pd.merge(merge_table, avg_reading_by_school["reading_score"].mean(), on="school_name")
merge_table

students_complete_grouped_by_school = students_complete.groupby(["school_name"])
total_students_count = students_complete_grouped_by_school.count()
total_students_count = total_students_count["math_score"]
total_students_count

math_passing_only_grouped_by_school = math_passing_only.groupby(["school_name"])
math_passing_students_count = math_passing_only_grouped_by_school.count()
math_passing_students_count = math_passing_students_count["math_score"]
math_passing_students_count

reading_passing_only_grouped_by_school = reading_passing_only.groupby(["school_name"])
reading_passing_students_count = reading_passing_only_grouped_by_school.count()
reading_passing_students_count = reading_passing_students_count["reading_score"]
reading_passing_students_count

merged_table_calcs = pd.merge(total_students_count, math_passing_students_count, on="school_name")
merged_table_calcs = pd.merge(merged_table_calcs, reading_passing_students_count, on="school_name")
merged_table_calcs = merged_table_calcs.rename(columns = {"math_score_x":"total students",
                                                         "math_score_y":"students passing math",
                                                         "reading_score":"students passing reading"})
merged_table_calcs["% passing math"] = merged_table_calcs["students passing math"]/merged_table_calcs["total students"]
merged_table_calcs["% passing reading"] = merged_table_calcs["students passing reading"]/merged_table_calcs["total students"]
merged_table_calcs = merged_table_calcs[["% passing math", "% passing reading"]]
merged_table_calcs

merge_table = pd.merge(merge_table, merged_table_calcs, on="school_name")
merge_table["overall passing rate"] = (merge_table["% passing math"] + merge_table["% passing reading"])/2
merge_table

merge_table = merge_table.rename(columns = {"size":"total students",
                                           "math_score":"avg math score",
                                           "reading_score":"avg reading score"})
merge_table
```

```python
### Top Performing Schools (By Passing Rate)

#* Create a table that highlights the top 5 performing schools based on Overall Passing Rate. Include:
#  * School Name
#  * School Type
#  * Total Students
#  * Total School Budget
#  * Per Student Budget
#  * Average Math Score
#  * Average Reading Score
#  * % Passing Math
#  * % Passing Reading
#  * Overall Passing Rate (Average of the above two)

top_schools = merge_table.sort_values(["overall passing rate"], ascending=False)
top_schools = top_schools.head(5)
top_schools
```

```python
bottom_schools = merge_table.sort_values(["overall passing rate"], ascending=True)
bottom_schools = bottom_schools.head(5)
bottom_schools
```

```python

### Math Scores by Grade\*\*

#* Create a table that lists the average Math Score for students of each grade level 
#(9th, 10th, 11th, 12th) at each school.
by_grade = students_complete.groupby(["grade"])
math_by_grade = by_grade["math_score"].mean()
math_by_grade
```

```python

### Reading Scores by Grade

#* Create a table that lists the average Reading Score
#for students of each grade level (9th, 10th, 11th, 12th) at each school.
reading_by_grade = by_grade["reading_score"].mean()
reading_by_grade
```

```python
### Scores by School Spending

#* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
#  * Average Math Score
#  * Average Reading Score
#  * % Passing Math
#  * % Passing Reading
#  * Overall Passing Rate (Average of the above two)

bins = [0,600,625,649,merge_table["budget_per_student"].max()]
group_names = ["<=600","601-625","626-649",">=650"]
merge_table["budget_bins"] = pd.cut(merge_table["budget_per_student"], bins, labels=group_names)

merge_table_1 = merge_table.groupby("budget_bins")
merge_table_1 = merge_table_1.mean()
merge_table_1 = merge_table_1[["avg math score", "avg reading score", "% passing math", "% passing reading", "overall passing rate"]]
merge_table_1
```

```python
### Scores by School Size

#* Repeat the above breakdown, but this time group schools based on a 
#reasonable approximation of school size (Small, Medium, Large).


bins = [0,1500,3000,4499,merge_table["total students"].max()]
group_names = ["<=1500","1501-3000","3001-4499",">=4500"]
merge_table["size_bins"] = pd.cut(merge_table["total students"], bins, labels=group_names)

merge_table2 = merge_table.groupby("size_bins")
merge_table2 = merge_table2.mean()
merge_table2 = merge_table2[["avg math score","avg reading score", "% passing math", "% passing reading", "overall passing rate" ]]
merge_table2
```

```python
### Scores by School Type

#* Repeat the above breakdown, but this time group schools based on school type (Charter vs. District).

school_type_gb = merge_table.groupby("type")
school_type_gb = school_type_gb.mean()
school_type_gb = school_type_gb[["avg math score","avg reading score", "% passing math", "% passing reading", "overall passing rate"]]
school_type_gb
```
